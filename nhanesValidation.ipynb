# Microsim NHANES Validation

The goal of this notebook is to illustrate how to run the NHANES validation simulations. 

First we will import some libraries and switch to the Microsim code directory.

```python
#we need the os module in order to change directory to the Microsim directory later
import os

#all the code needed to run the validation simulations is included in the validation.py file
#please see the validation.py file for more details
from microsim.validation import Validation

#change directory to the Microsim directory-this will be different for each user
#you will need to set microsimDir to be the directory where the Microsim code is in your computer
microsimDir = "/Users/deligkaris.1/OneDrive - The Ohio State University Wexner Medical Center/MICROSIM/CODE/microsim"
os.chdir(microsimDir)
```

Then, we will run all the validation code.
The code runs the steps to validate:
- baseline simulated population
- vascular risk factors over time
- CV event incidence and mortality
- dementia incidence
- blood pressure treatment effects

```python
#feel free to modify the sampleSize and nWorkers parameters
#sampleSize should be at least 0.5 million for good treatment effect estimates
#use a smaller sampleSize if you just want to see how this works, eg 10,000
#sampleSize currently affects only the size of the population used in the treatment effect simulations, not the rest
#nWorkers depends on how many cores you can use
Validation.nhanes(sampleSize=10000, nWorkers=4)
```
